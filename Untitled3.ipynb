```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("EV_Company_Vehicle_Types_1500.csv")

# Clean data columns
df['Vehicle_Type'] = df['Vehicle_Type'].str.strip().str.lower()
df['Battery_Capacity_kWh'] = pd.to_numeric(df['Battery_Capacity_kWh'], errors='coerce')

# Drop missing values
data_vb = df.dropna(subset=['Vehicle_Type', 'Battery_Capacity_kWh'])

# Encode categorical vehicle types numerically for clustering
vehicle_labels = data_vb['Vehicle_Type'].astype('category').cat.codes

# Prepare data for clustering
X_vb = pd.DataFrame({
    'VehicleLabel': vehicle_labels,
    'Battery_Capacity_kWh': data_vb['Battery_Capacity_kWh']
})

# Apply KMeans clustering
kmeans_vb = KMeans(n_clusters=3, random_state=42)
clusters_vb = kmeans_vb.fit_predict(X_vb)
data_vb['Cluster'] = clusters_vb

# Jitter on x-axis to avoid overlap
vehicle_categories = data_vb['Vehicle_Type'].astype('category')
vehicle_pos = vehicle_categories.cat.codes
jitter = np.random.uniform(-0.15, 0.15, size=len(vehicle_pos))

# Plot
plt.figure(figsize=(12, 7))
scatter = plt.scatter(vehicle_pos + jitter, data_vb['Battery_Capacity_kWh'],
                      c=data_vb['Cluster'], cmap='Set1', alpha=0.8, edgecolor='black')
plt.xticks(ticks=np.arange(len(vehicle_categories.cat.categories)),
           labels=vehicle_categories.cat.categories, rotation=45, ha='right')
plt.xlabel('Vehicle Type')
plt.ylabel('Battery Capacity (kWh)')
plt.title('Vehicle Type vs Battery Capacity')
plt.ylim(0, max(data_vb['Battery_Capacity_kWh'].max()+5, 40))
plt.colorbar(scatter, label='Cluster')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
df['Wheeler_Type'] = df['Wheeler_Type'].str.strip().str.lower()
df['Range_km'] = pd.to_numeric(df['Range_km'], errors='coerce')

data_wr = df.dropna(subset=['Wheeler_Type', 'Range_km'])

wheeler_labels = data_wr['Wheeler_Type'].astype('category').cat.codes

X_wr = pd.DataFrame({
    'WheelerLabel': wheeler_labels,
    'Range_km': data_wr['Range_km']
})

kmeans_wr = KMeans(n_clusters=3, random_state=42)
clusters_wr = kmeans_wr.fit_predict(X_wr)
data_wr['Cluster'] = clusters_wr

wheeler_categories = data_wr['Wheeler_Type'].astype('category')
wheeler_pos = wheeler_categories.cat.codes
jitter = np.random.uniform(-0.15, 0.15, size=len(wheeler_pos))

plt.figure(figsize=(12, 7))
scatter = plt.scatter(wheeler_pos + jitter, data_wr['Range_km'],
                      c=data_wr['Cluster'], cmap='Set2', alpha=0.8, edgecolor='black')
plt.xticks(ticks=np.arange(len(wheeler_categories.cat.categories)),
           labels=wheeler_categories.cat.categories, rotation=45, ha='right')
plt.xlabel('Wheeler Type')
plt.ylabel('Range (km)')
plt.title('Segmentation: Wheeler Type vs Range')
plt.colorbar(scatter, label='Cluster')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
df['Price_INR_Lakhs'] = pd.to_numeric(df['Price_INR_Lakhs'], errors='coerce')
data_wp = df.dropna(subset=['Wheeler_Type', 'Price_INR_Lakhs'])

wheeler_labels_wp = data_wp['Wheeler_Type'].astype('category').cat.codes

X_wp = pd.DataFrame({
    'WheelerLabel': wheeler_labels_wp,
    'Price_INR_Lakhs': data_wp['Price_INR_Lakhs']
})

kmeans_wp = KMeans(n_clusters=3, random_state=42)
clusters_wp = kmeans_wp.fit_predict(X_wp)
data_wp = data_wp.copy()
data_wp['Cluster'] = clusters_wp

# Jitter to avoid overlap
wheeler_categories = data_wp['Wheeler_Type'].astype('category')
wheeler_pos = wheeler_categories.cat.codes
jitter = np.random.uniform(-0.15, 0.15, size=len(wheeler_pos))

plt.figure(figsize=(12, 7))
scatter = plt.scatter(wheeler_pos + jitter, data_wp['Price_INR_Lakhs'],
                      c=data_wp['Cluster'], cmap='Set3', alpha=0.8, edgecolor='black')
plt.xticks(ticks=np.arange(len(wheeler_categories.cat.categories)),
           labels=wheeler_categories.cat.categories, rotation=45, ha='right')
plt.xlabel('Wheeler Type')
plt.ylabel('Price (INR Lakhs)')
plt.title('Segmentation: Wheeler Type vs Price')
plt.colorbar(scatter, label='Cluster')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("EV_Company_Vehicle_Types_1500.csv")

# General overview
summary = df.describe(include='all')

# Count of unique vehicle types
vehicle_type_counts = df['Vehicle_Type'].value_counts()

# Count of unique wheeler types
wheeler_type_counts = df['Wheeler_Type'].value_counts()

# Aggregated stats by vehicle type
agg_vehicle = df.groupby('Vehicle_Type').agg({
    'Price_INR_Lakhs': ['mean', 'min', 'max'],
    'Range_km': ['mean', 'min', 'max'],
    'Battery_Capacity_kWh': ['mean', 'min', 'max'],
    'Top_Speed_kmph': ['mean', 'min', 'max'],
    'Model_Name': 'count'
}).rename(columns={'Model_Name': 'Count'})

# Insights dictionary
insights = {
    'Most Common Vehicle Types': vehicle_type_counts.head(5).to_dict(),
    'Most Common Wheeler Types': wheeler_type_counts.head(5).to_dict(),
    'Battery Capacity Range (kWh)': (df['Battery_Capacity_kWh'].min(), df['Battery_Capacity_kWh'].max()),
    'Price Range (INR Lakhs)': (df['Price_INR_Lakhs'].min(), df['Price_INR_Lakhs'].max()),
    'Range km': (df['Range_km'].min(), df['Range_km'].max()),
    'Top Speed (kmph)': (df['Top_Speed_kmph'].min(), df['Top_Speed_kmph'].max()),
    'Aggregated Stats By Vehicle Type': agg_vehicle
}

insights
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Load filtered dataset
df = pd.read_csv("EV_Company_Vehicle_Types_1500.csv")

# Clean and prepare data
df['Price_INR_Lakhs'] = pd.to_numeric(df['Price_INR_Lakhs'], errors='coerce')
df['Range_km'] = pd.to_numeric(df['Range_km'], errors='coerce')

data_pr = df.dropna(subset=['Price_INR_Lakhs', 'Range_km'])

X_pr = data_pr[['Price_INR_Lakhs', 'Range_km']]

# Apply KMeans clustering for 3 clusters
kmeans_pr = KMeans(n_clusters=3, random_state=42)
clusters_pr = kmeans_pr.fit_predict(X_pr)

data_pr['Cluster'] = clusters_pr

# Plot scatter with clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_pr, x='Price_INR_Lakhs', y='Range_km', hue='Cluster', palette='Set2')
plt.title('Segmentation: Price vs Range')
plt.xlabel('Price (INR Lakhs)')
plt.ylabel('Range (km)')
plt.legend(title='Cluster')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
